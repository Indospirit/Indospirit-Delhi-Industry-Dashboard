```python
def generate_table(dfx, max_rows=50):
    return html.Table(
        # Header
        [html.Tr([html.Th(col) for col in dfx.columns])] +

        # Body
        [html.Tr([
            html.Td(dfx.iloc[i][col]) for col in dfx.columns
        ]) for i in range(min(len(dfx), max_rows))]
    )
```

```python
available_brands1 = []
available_brands2 = []
available_category = []
available_industry = []
available_types = []
available_vends = []
available_sr = []

available_brands1 = np.append(available_brands1,'ALL')
available_brands2 = np.append(available_brands2,'ALL')
available_category = np.append(available_category,'ALL')
available_industry = np.append(available_industry,'ALL')
available_types = np.append(available_types,'ALL')
available_vends = np.append(available_vends,'ALL')



available_brands1 = np.append(available_brands1,df4['Brand.Name'].unique())
available_brands2 = np.append(available_brands2,df4['Brand.Code'].unique())
available_types = np.append(available_types,df4['ON_OFF'].unique())
available_industry = np.append(available_industry,df4['Industry'].unique())
available_category = np.append(available_category,df4['Liquor.Category'].unique())
available_vends = np.append(available_vends,df4['Deptt'].unique())
available_sr = np.append(available_sr,df3['Sales Rep name'].unique())


VALID_USERNAME_PASSWORD_PAIRS = [
    ['AB GRAIN SPIRITS PVT. LTD.', 'admin']
]



#df4 = df4[df4['Supplier.Name'] == username]

### starting of app making ###
### app authentication system ###
app = dash.Dash('auth')
auth = dash_auth.BasicAuth(
    app,
    VALID_USERNAME_PASSWORD_PAIRS
)

app.css.append_css({'external_url': 'https://cdn.rawgit.com/plotly/dash-app-stylesheets/2d266c578d2a6e8850ebce48fdb52759b2aef506/stylesheet-oil-and-gas.css'})
app.css.append_css({'external_url':'my_stle.css'})
# Loading screen CSS
app.css.append_css({"external_url": "https://codepen.io/chriddyp/pen/brPBPO.css"})


### app body starts here ###

app.layout = html.Div([

html.Div(id = 'main-datframe', style={'display': 'none'}),

#### A floating tab on screen to keep in chedck all the chosen parameters ####
html.Div([
    html.Table(id='Tab-keeper',),
],className='row',style={'position':'fixed','width':'200px','right':'0','top':'0','float':'right','color':'#ffffff','background-color':'#001625','overflow':'auto'},),

html.Div(
            [
                
                html.Img(
                    src="https://static.wixstatic.com/media/c956b6_56e1c0a0f0d44f579a3c9e7ebe5ba8de~mv2_d_3159_1999_s_2.png/v1/fill/w_71,h_44,al_c,usm_0.66_1.00_0.01/c956b6_56e1c0a0f0d44f579a3c9e7ebe5ba8de~mv2_d_3159_1999_s_2.png",
                    className='two columns',
                     style={'float': 'left', 'position': 'relative','width':'85px','margin-left':'125px','padding-top':'15px',
                         
                     },
                ),
                html.H1(
                    'Indospirit-Delhi Industry Dashboard',
                    className='eight columns',
                    style = {'color':'#ffffff','text-align':'center','font-size':'35px','padding':'20px 0px 20px'}
 
                ),
            ],
            className='row',style={"background-color":"#001652","height":"75px","width":"100%",
                                   "margin-left":"-0.5%","margin-right":"-2%"}
        ),
    
html.Div([
    html.H6('Year Select'),
    dcc.Slider(
    min=2014,
    max=dt.now().year,
    marks={i: '{}'.format(i) for i in range(2014,dt.now().year+1)},
    value=dt.now().year,
    id = 'Year-picker'
)
],className='row',style={'width': '30%','margin':'80 80 80 80','padding':'20px 50px 0px 25px'}),   

html.Div([
    ### Month select slider ###
            html.H6('Months Select'),
            dcc.RangeSlider(
                min=1,
                max=12,
                step=None,
                allowCross = False,
                marks=month_range,
                id = 'Month-Picker',
                value=[1, 4]),
            ],style={'float': 'centre','width': '60%','margin':'80 80 80 80','padding':'20px 50px 0px 25px'}),
html.Div([
    html.Button(
        'Submit',
        id = 'Main-Button',)
],className='row',style={'margin-left':'500px','margin-top':'50px'}),    

html.Div([   
html.Div([
    ### On and OFF dropdown body ###
            html.H6('Channel Type(CT)',),
            dcc.Dropdown(
                id='ON-OFF',
                options=[{'label': i, 'value': i} for i in available_types],
                value='ALL',
                clearable=False,
            ),],className='two columns',style={'width': '20%', 'float': 'centre','margin':'80 80 80 80',
                                               'padding':'20px 50px 0px 25px'}),
html.Div([
    ### Vendor type dropdown body ###
             html.H6('Retail Type(RT)',),
            dcc.Dropdown(
                id='vend-type',
                value=available_vends,
                clearable=False,
                multi = True
            ),],className='seven columns',style={'width': '50%','float': 'centre','margin':'80 80 80 80',
                                                 'padding':'20px 50px 0px 25px'}),
],className='row'),

html.Div([
    ### Price range dropdown body ###
            html.H6('Price Select(PR)',),
              dcc.Input(
              placeholder = 'Min Price',
              type = 'number',
              value = 0,
              inputmode = 'numeric',
              id = 'min-price'),
    
              dcc.Input(
              placeholder = 'Max Price',
              value = 300000,
              type = 'number',
              inputmode = 'numeric',
              id = 'max-price',),
            ],style={'float': 'centre','width': '80%','margin':'80 80 80 80','padding':'20px 50px 0px 25px'}),
html.Div([
    ### IFL and IMFL dropdown body ###
            html.H6('Industry Type(IS)',),
            dcc.Dropdown(
                id='Industry',
                options=[{'label': i, 'value': i} for i in available_industry],
                value='ALL',
                clearable=False,
            ),],className='row',style={'width': '70%', 'float': 'centre','margin':'80 80 80 80',
                                       'padding':'20px 50px 0px 25px'}),
html.Div([
html.Div([
    ### Category of liquor dropdown body ###
            html.H6('Liquor Category(LC)',),
            dcc.Dropdown(
                id='Category',
                options=[{'label': i, 'value': i} for i in available_category],
                value='ALL',
                clearable=False,
            ),
            
            ],className = 'three columns',style={'width': '30%', 'float': 'centre','margin':'80 80 80 80',
                                                 'padding':'20px 50px 0px 25px'}),                         
html.Div([
    ### Brand Code dropdown body ###
            html.H6('Brand Code(BC)',),
            dcc.Dropdown(
                id='Brand-Code',
                value='ALL',
                clearable=False,
            ),],className='three columns',style={'width': '30%', 'float': 'centre','margin':'80 80 80 80',
                                                 'padding':'20px 50px 0px 25px'}),
html.Div([
    ### Brand Name ####
        html.H6('Brand Name',),
        html.Table(id = 'Brand-Name'),
              ],className='three columns',style={'width': '30%', 'float': 'centre','margin':'80 80 80 80',
                                                 'padding':'20px 50px 0px 25px'})
],className='row'),

html.Div([
    html.Div([
    ### Sales Rep dropdown body ###
             html.H6('Sales Rep Selection(SR)',),
                dcc.Dropdown(
                    id='sales-rep',
                    options=[{'label': i, 'value': i} for i in available_sr],
                    value=available_sr,
                    clearable=False,
                    multi = True
                ),],className='six columns',style={'width': '50%','float': 'centre','margin':'80 80 80 80',
                                                 'padding':'20px 50px 0px 25px'}),
],className = 'row'),

### Map body starts here ###
html.Div([
    #html.H6('Delhi Distribution(MC)(CT-PR-RT-SR)'),
dcc.Graph(
    className = 'eight columns',
    id= 'indicator-graphic',
    config = dict(scrollZoom = True , mapboxAccessToken = mapbox_access_token , autosizable = True ,
                  frameMargins = dict(l=0,r=0,b=0,t=0,pad=20)) , 
        ),
html.Div(
    
    [
        html.H5('Shop sales(CT-PR-RT-SR)'),
        html.Table(id = 'Shop-table',)
        
    ],style={'padding': '50','float':'Center','height':'500','overflow':'auto'},className = 'four columns'),


    ],style={'float':'Center','padding': '50'}, 
    className = "row"),
### Map body ends here ###
html.Div([
    html.Button(
        'Reset the shop code',
        id = 'reset-button-1',)
],className='row',style={'float':'left','margin-left':'500px','margin-top':'-50px'}),    



html.Div([
    ### Pie chart body starts here ###
html.Div([
    html.H6('Brand wise liquor market distribution(LC-IS)'),
    dcc.Graph(
    #figure = fig_pie,
    id = 'Pie-chart'),
],style={'float':'Left'},className = 'six columns'),
### Pie chart body ends here ###


### Regression lines plot start here ###
html.Div([
dcc.Graph(
    #figure = fig_reg,
    id = 'Regression lines for products',
)
],style={'plot_bgcolor':'#14307e','paper_bgcolor':'#14307e','float':'Center'},
    className = "six columns"),
### Regression lines plot ends here ###
],className = 'row'),

    
    
html.Div([

    html.Div([
        html.H6('Category v/s Department performance(PR-IS)'),
    ### Heatmap ###
        dcc.Graph(
            id = 'Heatmap',
            #figure = fig_heatmap,
        ),
    ],style={'float':'Center'},className = "six columns"),
    
    html.Div([
    ### Heatmap generated table ###
        html.H6('Brands sold in the particular category and department(PR-IS)'),
        html.Table(id = 'Heatmap-pie'),
    ],style={'float':'Center','height':'400px','overflow':'auto'},className = 'six columns'),

],className = 'row'),
    


html.Details([
    html.Summary('A Summary table view'),
    html.Div(
    
    [
        html.H5('Table Generated for Summary'),       
        html.Div([
            #### Table Generation ####
            html.H5('Monthly Retailer sales(CT-RT-PR-IS-LC-BC-SR)'),
            html.Table(id = 'Table-Gen-Sum1',),
        ],className='four columns',style={'padding': '50','margin': '100 0 100 50','display':'inline-Block'}),
        html.Div([
            #### Table Generation ####
            html.H5('Top Selling Brands(CT-PR-IS-LC)'),
            html.Table(id = 'Table-Gen-Sum2')
        ],className='four columns',style={'padding': '50','margin': '100 0 100 50','display':'inline-Block'}),
        html.Table(id = 'Own-Brands',
                   className='four columns',
#                    style={'padding': '50', 
#                           'margin': '100 0 100 50',
#                           'display':'inline-Block',}
                         
                  ),
        
    ],className = 'row'),
]),
    
    
html.Details([
    html.Summary('A Pivot table view'),
    html.Div(
    
    [
        #### Download button is generated ####
        html.H5('Table Generated with Groupby(MC-CT-RT-PR-IS-LC-BC-SR)'),
        html.A(
            'Click on this to download the Data',
            id='download-link-groupby',
            download="Filtered_Data.csv",
            href="",
            target="_blank",
            style={'float': 'left'}
        ),
        
        #### Table Generation ####
        html.Table(id = 'Table-Gen-Groupby',),
        
    ],style={'padding': '50', 'margin': '100 100 100 100'}
),
]),
    
    

html.Details([
    html.Summary('A total table view'),
    html.Div(
    
    [
        #### Download button is generated ####
        html.H5('Table Generated(MC-CT-RT-PR-IS-LC-BC-SR)'),
        html.A(
            'Click on this to download the Data',
            id='download-link',
            download="Filtered_Data.csv",
            href="",
            target="_blank",
            style={'float': 'left'}
        ),
        
        #### Table Generation ####
        html.Table(id = 'Table Gen',),
        
    ],style={'padding': '50', 'margin': '100 100 100 100'}
),
]),

    
],style={'float': 'center' , 'margin': '100 200 100 200'})

### app layout body ends here ###

# @app.callback(
#     dash.dependencies.Output('main-dataframe','children'),
#     [dash.dependencies.State('Year-picker','value'),
#      dash.dependencies.State('Month-picker','value'),
#     ],
#     [dash.dependencies.Input('Main-button','n_clicks'),
#     ],
# )
# def dataframe_creater(selected_year,selected_months,submited):
#     df4 = pd.read_sql()
#       if flask.request.headers.get('Authorization', None) is not None:
#         username_password = base64.b64decode(header.split('Basic ')[1])
#         username_password_utf8 = username_password.decode('utf-8')
#         username, password = username_password_utf8.split(':')
#         df4 = df4[df4['Supplier.Name'] == username]
#     return df4


###### custom function to filter databases on basis of price filter ######
##########################  Body starts here  ############################
def filter_price(df,value_min,value_max):
    return df[(df['MRP..INR.'] > value_min) & (df['MRP..INR.'] < value_max)]
        
##############  Body ends here  #################

###### app callback to update the clickdata to null  ######
################  Body starts here  ######################
@app.callback(
    dash.dependencies.Output('indicator-graphic', 'clickData'),
    [dash.dependencies.Input('reset-button-1', 'n_clicks'),
    ]
)
def reset_function(clicked_button):
    return None
    
################   Body ends here   ######################

###### custom function to filter databases ######
#############  Body starts here  ################
def filter_data(df,col_name,value):
    if value == 'ALL':
        return df
    else:
        return df[df[col_name] == value]
##############  Body ends here  #################



###### custom function to filter databases on basis of radio buttons######
#############               Body starts here              ################
def filter_radio(df,col_name,value):
    if value == 'ALL':
        return df
    else:
        return df[df[col_name].isin(value)]
##############  Body ends here  #################


#### app callback to update the tabkeeper ####
#### Body starts here ####
@app.callback(
    dash.dependencies.Output('Tab-keeper','children'),
    [#dash.dependencies.Input('main-dataframe','children'),
     dash.dependencies.Input('Industry', 'value'),
     dash.dependencies.Input('Category', 'value'),
     dash.dependencies.Input('Brand-Code', 'value'),
     dash.dependencies.Input('ON-OFF', 'value'),
     #dash.dependencies.Input('vend-type', 'value'),
     dash.dependencies.Input('indicator-graphic', 'clickData'),
     dash.dependencies.Input('min-price', 'value'),
     dash.dependencies.Input('max-price', 'value'),
    ]
)
def tab_changer(selected_indus,selected_cat,selected_b_code,selected_v_type,clicked_data,selected_min_price,selected_max_price):
    if clicked_data is not None:
        z = clicked_data['points'][0]['customdata']
    else:
        z = 'ALL'
    dff = pd.DataFrame(data = [{'Industry(IS)':selected_indus,
                               'Category(LC)':selected_cat,
                               'BrandCode(BC)':selected_b_code,
                               'ON/OFF(CT)':selected_v_type,
                               'ShopCode(MC)':z,
                               'LowerPrice(PR-min)':selected_min_price,
                               'UpperPrice(PR-max)':selected_max_price,}])

        
    return(generate_table(dff))
#### Body ends here ####


#### app callback to update dropdown for brand code   ####
################  Body starts here  ######################

@app.callback(
    dash.dependencies.Output('Brand-Code', 'options'),
    [#dash.dependencies.Input('main-dataframe','children'),
     dash.dependencies.Input('Category', 'value'),
     dash.dependencies.Input('Industry', 'value'),
     dash.dependencies.Input('min-price' , 'value'),
     dash.dependencies.Input('max-price' , 'value'),
    ]
)
def update_brand_code(selected_cat,selected_indus,selected_min_price,selected_max_price):
    dff = filter_data(df4,'Liquor.Category',selected_cat)           #### Filter on the basis of product category
    dff = filter_data(dff,'Industry',selected_indus)                #### Filter on the basis of industry type
    dff = filter_price(dff,selected_min_price,selected_max_price)   #### Filter on the basis of price
    x = ['ALL']
    x = np.append(x,dff['Brand.Code'].unique())
    option = [{'label': i, 'value': i} for i in x]
    return(option)

################   Body ends here   ######################

#### app callback to update dropdown for retailer type ####
################  Body starts here  ######################

@app.callback(
    dash.dependencies.Output('vend-type', 'options'),
    [#dash.dependencies.Input('main-dataframe','children'),
     dash.dependencies.Input('ON-OFF', 'value'),
    ]
)
def update_vendor(selected_v_type):
    dff = filter_data(df4,'ON_OFF',selected_v_type)                 #### Filter on the ON and OFF type
    x = []
    #x = np.append(x,'ALL')
    x = np.append(x,dff['Deptt'].unique())
    option = [{'label': i, 'value': i} for i in x]
    return(option)

################   Body ends here   ######################

#### app callback to update the Brand name generated  ####
################  Body starts here  ######################
@app.callback(
    dash.dependencies.Output('Brand-Name','children'),
    [dash.dependencies.Input('Brand-Code', 'value'),],
)
def brand_name_gen(selected_b_code):
    if selected_b_code != 'ALL':
        v = df2_1[df2_1['Brand Code'] == selected_b_code]
        return(v['Product_Name'].unique())
################  Body ends here  ######################
    
###### app callback to update the mapbox generated  ######
################  Body starts here  ######################
    
@app.callback(
    dash.dependencies.Output('indicator-graphic', 'figure'),
    [#dash.dependencies.Input('main-dataframe','children'),
     dash.dependencies.Input('ON-OFF', 'value'),
     dash.dependencies.Input('vend-type', 'value'),
     dash.dependencies.Input('min-price' , 'value'),
     dash.dependencies.Input('max-price' , 'value'),
     dash.dependencies.Input('sales-rep','value'),
    ]
)
def change_main_graph(selector1,selector2,selector3,selector4,selector5):                             
    dff = filter_data(df4,'ON_OFF',selector1)                            #### Filter on the basis of vend type
    dff = filter_radio(dff,'Deptt',selector2)                            #### Filter on the basis of vendor
    dff = filter_price(dff,selector3,selector4)                          #### Filter on the basis of price
    dffx = filter_radio(df3,'Sales Rep name',selector5)                  #### Filter the selected Sales Reps only
    dff = dffx[dffx['Shop Code'].isin(dff['Shop.Code'])]                 #### Filter on the shop codes present in selected vends
    traces = []
    for sr, dfff in dff.groupby('Sales Rep name'):
        trace = dict(
            type='scattermapbox',
            lon=dfff['longitude'],
            lat=dfff['latitude'],
            text=dfff['name'],
            customdata=dfff['Shop Code'],
            hoverinfo = dfff['name']+':'+dfff['Sales Rep name'],
            name=sr,
            marker=dict(
                size=8,
                opacity=0.6
            )
        )
        traces.append(trace)
    figure = dict(data=traces, layout=layout1)
    return(figure)
################   Body ends here   ######################

#### app callback to update the table shop generated  ####
################  Body starts here  ######################
@app.callback(
    dash.dependencies.Output('Shop-table', 'children'),
    [#dash.dependencies.Input('main-dataframe','children'),
     dash.dependencies.Input('indicator-graphic', 'clickData'),
     dash.dependencies.Input('min-price' , 'value'),
     dash.dependencies.Input('max-price' , 'value'),
    ]
)
def update_shop(clicked_data,selected_min_price,selected_max_price):
    if clicked_data is not None:
        z = clicked_data['points'][0]['customdata']
        x_df = filter_data(df4,'Shop.Code',z)
        x_df = filter_price(x_df,selected_min_price,selected_max_price)
        x_df = x_df.drop(labels=['S.no.', 'Industry', 'Purchase.Order.No.', 'PO.Date', 'Supplier.Name','Supplier.Code',
                               'Transport.Permit.No.', 'TP.Status', 'TP.Date','Shop.Code','Liquor.Sub.Category',
                                'Brand.Code','Size..ml.', 'Packing','WSP.per.Unit..INR.',
                               'WSP.per.Case..INR.','Excise.Revenue.per.Unit..INR.', 'MRP..INR.',
                               'WSP.per.Unit...No.of.Cases.Packing.Size','Excise.revenue.Per.Unit...No.Of.Cases..Packing.Size',
                               'Deptt', 'Month','ON_OFF','Brand.Name' ], axis = 1)
        x_df = x_df.groupby(by = ['Shop.Name','Liquor.Category']).sum()
        x_df = x_df.reset_index()
        x_df = x_df.sort_values(by = ['Quantity.Ordered'],ascending=False)
        return(generate_table(x_df))
################  Body ends here  ######################


#### app callback to update the Pie chart generated  ####
################  Body starts here  ######################
    
@app.callback(
    dash.dependencies.Output('Pie-chart', 'figure'),
    [#dash.dependencies.Input('main-dataframe','children'),
     dash.dependencies.Input('Category', 'value'),
     dash.dependencies.Input('Industry','value'),
    ]
)
def update_Pie(selected_cat,selected_indus):
    x_df = filter_data(df4,'Liquor.Category',selected_cat)
    x_df = filter_data(x_df,'Industry',selected_indus)
    x_df = x_df.drop(labels=['S.no.', 'Industry', 'Purchase.Order.No.', 'PO.Date', 'Supplier.Name','Supplier.Code',
                               'Transport.Permit.No.', 'TP.Status', 'TP.Date','Shop.Code', 'Shop.Name', 'Liquor.Category',
                               'Liquor.Sub.Category','Brand.Code','Size..ml.', 'Packing','WSP.per.Unit..INR.',
                               'WSP.per.Case..INR.','Excise.Revenue.per.Unit..INR.', 'MRP..INR.',
                               'WSP.per.Unit...No.of.Cases.Packing.Size','Excise.revenue.Per.Unit...No.Of.Cases..Packing.Size',
                               'Deptt', 'Month','ON_OFF' ], axis = 1)
    x_df = x_df.groupby(by = ['Brand.Name']).sum()
    x_df = x_df.reset_index()
    x_df = x_df.sort_values(by = ['Quantity.Ordered'],ascending=False)
    top_five = x_df.head(5)
    label = ['Other']
    values = [x_df['Quantity.Ordered'].sum() - top_five['Quantity.Ordered'].sum()]
    for i in range(0,len(top_five)):
        label = np.append(label,top_five['Brand.Name'].iloc[i])
        values = np.append(values,top_five['Quantity.Ordered'].iloc[i])

    fig_pie = {
        "data": [
        {
            "values": values,
            "labels": label,
            "domain": {"x": [0, .68]},
            "name": "Sales of the brands",
            "hoverinfo":"label+percent+name",
            "hole": .4,
            "type": "pie"
        }],
          "layout": {
            "title":"Category sales in chosen period",
            "legend": {
              "x": 0.78, 
              "y": 0.95, 
              "borderwidth": 0, 
              "font": {
                    "family": "Arial", 
                    "size": 9
                      }, 
              "orientation": "v", 
              "traceorder": "normal"
              },
            "annotations": [
                {
                    "font": {
                        "size": 20
                    },
                    "showarrow": False,
                    "text": selected_cat,
                    "x": 0.35,
                    "y": 0.5
                }
            ]
        }
    }
    return fig_pie
##############  Body ends here  #################

 
##### app callback to create the line graph plot #####
################  Body starts here  ######################
@app.callback(
    dash.dependencies.Output('Regression lines for products','figure'),
    [#dash.dependencies.Input('main-dataframe','children'),
     dash.dependencies.Input('Category', 'value'),],
)
def update_line_plot(selected_cat):
    beer_df = df4[df4['Liquor.Category']=='Beer']
    whisky_df = df4[df4['Liquor.Category']=='Whisky']
    wine_df = df4[df4['Liquor.Category']=='Wine']
    vodka_df = df4[df4['Liquor.Category']=='Vodka']
    rum_df = df4[df4['Liquor.Category']=='Rum']
    d_beer_df = df4[df4['Liquor.Category']=='Draught Beer']
    gin_df = df4[df4['Liquor.Category']=='Gin']
    alco_df = df4[df4['Liquor.Category']=='Alcopop']
    mab_df = df4[df4['Liquor.Category']=='Mixed Alcoholic Beverages']
    liq_df = df4[df4['Liquor.Category']=='Liqueur']
    brandy_df = df4[df4['Liquor.Category']=='Brandy']



    time_line = []
    beer_y = []
    wine_y = []
    whisky_y = []
    rum_y = []
    vodka_y = []
    dbeer_y = []
    gin_y = []
    alc_y = []
    mab_y = []
    liq_y = []
    brand_y = []

    for i in (df4['Month'].unique()):
        time_line = np.append(time_line,i)
        beer_y = np.append(beer_y,sum(beer_df[beer_df['Month']== i]['Quantity.Ordered']))
        wine_y = np.append(wine_y,sum(wine_df[wine_df['Month']== i]['Quantity.Ordered']))
        whisky_y = np.append(whisky_y,sum(whisky_df[whisky_df['Month']== i]['Quantity.Ordered']))
        rum_y = np.append(rum_y,sum(rum_df[rum_df['Month']== i]['Quantity.Ordered']))
        vodka_y = np.append(vodka_y,sum(vodka_df[vodka_df['Month']== i]['Quantity.Ordered']))
        dbeer_y = np.append(dbeer_y,sum(d_beer_df[d_beer_df['Month']== i]['Quantity.Ordered']))
        gin_y = np.append(gin_y,sum(gin_df[gin_df['Month']== i]['Quantity.Ordered']))
        alc_y = np.append(alc_y,sum(alco_df[alco_df['Month']== i]['Quantity.Ordered']))
        mab_y = np.append(mab_y,sum(mab_df[mab_df['Month']== i]['Quantity.Ordered']))
        liq_y = np.append(liq_y,sum(liq_df[liq_df['Month']== i]['Quantity.Ordered']))
        brand_y = np.append(brand_y,sum(brandy_df[brandy_df['Month']== i]['Quantity.Ordered']))


    trace_beer = []
    trace_wine = []
    trace_whisky = []
    trace_rum = []
    trace_vodka = []
    trace_d_beer = []
    trace_gin = []
    trace_alco = []
    trace_mab = []
    trace_liq = []
    trace_brandy = []

    trace_beer = go.Scatter(
        x = time_line,
        y = beer_y,
        mode = 'lines+markers',
        name = 'Beer',
    )
                 
    trace_wine = go.Scatter(
        x = time_line,
        y = wine_y,
        mode = 'lines+markers',
        name = 'Wine'
    )
                 
    trace_whisky = go.Scatter(
        x = time_line,
        y = whisky_y,
        mode = 'lines+markers',
        name = 'Whisky',
    )
                   
    trace_rum = go.Scatter(
        x = time_line,
        y = rum_y,
        mode = 'lines+markers',
        name = 'Rum',
    )
                
    trace_vodka = go.Scatter(
        x = time_line,
        y = vodka_y,
        mode = 'lines+markers',
        name = 'Vodka'
    )
                  
    trace_d_beer = go.Scatter(
        x = time_line,
        y = dbeer_y,
        mode = 'lines+markers',
        name = 'Draught Beer'
    )
                   
    trace_gin = go.Scatter(
        x = time_line,
        y = gin_y,
        mode = 'lines+markers',
        name = 'Gin'
    )
                
    trace_alco = go.Scatter(
        x = time_line,
        y = alc_y,
        mode = 'lines+markers',
        name = 'Alcopop'
    )
                 
    trace_mab = go.Scatter(
        x = time_line,
        y = mab_y,
        mode = 'lines+markers',
        name = 'Multiple Alcoholic Beverage'
    )
                
    trace_liq = go.Scatter(
        x = time_line,
        y = liq_y,
        mode = 'lines+markers',
        name = 'Liqueur'
    )

    trace_brandy = go.Scatter(
        x = time_line,
        y = brand_y,
        mode = 'lines+markers',
        name = 'Brandy'
    )
                   


    data = [trace_beer,
            trace_wine,
            trace_whisky,
            trace_rum,
            trace_alco,
            trace_brandy,
            trace_d_beer,
            trace_gin,
            trace_liq,
            trace_mab,
            trace_vodka]

    fig_reg = dict(data=data)
    return fig_reg
################  Body ends here  ######################


###### app callback to update the heatmap generated ######
################  Body starts here  ######################
@app.callback(
    dash.dependencies.Output('Heatmap','figure'),
    [#dash.dependencies.Input('main-dataframe','children'),
     dash.dependencies.Input('Industry', 'value'),
     dash.dependencies.Input('min-price' , 'value'),
     dash.dependencies.Input('max-price' , 'value'),
    ]
)
def update_heatmap(selected_indus,selected_min_price,selected_max_price):
    xl = filter_data(df4,'Industry',selected_indus)
    xl = filter_price(xl,selected_min_price,selected_max_price)
    xl = xl.drop(columns = ['S.no.', 'Industry', 'Purchase.Order.No.', 'PO.Date', 'Supplier.Name','Supplier.Code',
                            'Transport.Permit.No.', 'TP.Status', 'TP.Date','Shop.Code', 'Shop.Name','Liquor.Sub.Category',
                            'Brand.Code', 'Brand.Name', 'Size..ml.', 'Packing','WSP.per.Unit..INR.', 'WSP.per.Case..INR.',
                            'Excise.Revenue.per.Unit..INR.', 'MRP..INR.','WSP.per.Unit...No.of.Cases.Packing.Size',
                            'Excise.revenue.Per.Unit...No.Of.Cases..Packing.Size','Month','ON_OFF'])
    xl = xl.groupby(by = ['Liquor.Category','Deptt']).sum()
    xl = xl.reset_index()

    xk = pd.pivot_table(data = xl,aggfunc=np.sum,values='Quantity.Ordered',columns='Liquor.Category',index='Deptt')

    xz = xk.values
    trace_h_m = {
        'z' : xz,
        'y' : xk.index,
        'x' : xk.columns,
        'type': 'heatmap',
        'reversescale': True,
        'customdata': ['y','x'],
        'connectgaps': False,
        'xgap':2,
        'ygap':2,
        'colorscale':[
        [0, "rgb(252, 234, 172)"], [0.3, "rgb(254, 185, 127)"], [0.4, "rgb(250, 129, 94)"], [0.5, "rgb(176, 52, 122)"], [0.65, "rgb(98, 24, 127)"], [1, "rgb(4, 4, 21)"]],
    }
    data = [trace_h_m]
    fig_heatmap = {'data':data}
    return fig_heatmap    
################   Body ends here   ######################

#### app callback to update the heatmap Pie chart generated ####
###################  Body starts here  #########################
    
@app.callback(
    dash.dependencies.Output('Heatmap-pie', 'children'),
    [#dash.dependencies.Input('main-dataframe','children'),
     dash.dependencies.Input('Heatmap', 'clickData'),
     dash.dependencies.Input('Industry', 'value'),
     dash.dependencies.Input('min-price' , 'value'),
     dash.dependencies.Input('max-price' , 'value'),
    ]
)
def update_Heatmap_Pie(selected_cell,selected_indus,selected_min_price,selected_max_price):
    if selected_cell is not None:
        x_df = filter_data(df4,'Liquor.Category',selected_cell['points'][0]['x'])
        x_df = filter_data(x_df,'Deptt',selected_cell['points'][0]['y'])
        x_df = filter_data(x_df,'Industry',selected_indus)
        x_df = filter_price(x_df,selected_min_price,selected_max_price)
        x_df = x_df.drop(labels=['S.no.', 'Industry', 'Purchase.Order.No.', 'PO.Date', 'Supplier.Name','Supplier.Code',
                               'Transport.Permit.No.', 'TP.Status', 'TP.Date','Shop.Code', 'Shop.Name', 'Liquor.Category',
                               'Liquor.Sub.Category','Brand.Code','Size..ml.', 'Packing','WSP.per.Unit..INR.',
                               'WSP.per.Case..INR.','Excise.Revenue.per.Unit..INR.', 'MRP..INR.',
                               'WSP.per.Unit...No.of.Cases.Packing.Size','Excise.revenue.Per.Unit...No.Of.Cases..Packing.Size',
                               'Deptt', 'Month','ON_OFF' ], axis = 1)
        x_df = x_df.groupby(by = ['Brand.Name']).sum()
        x_df = x_df.reset_index()
        x_df = x_df.sort_values(by = ['Quantity.Ordered'],ascending=False)
        top_five = x_df.head(5)
        label = ['Other']
        values = [x_df['Quantity.Ordered'].sum() - top_five['Quantity.Ordered'].sum()]
        for i in range(0,len(top_five)):
            label = np.append(label,top_five['Brand.Name'].iloc[i])
            values = np.append(values,top_five['Quantity.Ordered'].iloc[i])
        test_dict = {'Brands': label,'Quantity Ordered': values}
        dff = pd.DataFrame.from_dict(test_dict)
        dff = dff.sort_values(by = ['Quantity Ordered'] ,ascending=False)
        return generate_table(dff)
###################  Body ends here  #########################


##  app callback to update the Sum1 to be tabulated  ##
################  Body starts here  ######################
@app.callback(
    dash.dependencies.Output('Table-Gen-Sum1', 'children'),
    [#dash.dependencies.Input('main-dataframe','children'),
     dash.dependencies.Input('ON-OFF' , 'value'),
     dash.dependencies.Input('Category', 'value'),
     dash.dependencies.Input('Brand-Code', 'value'),
     dash.dependencies.Input('min-price' , 'value'),
     dash.dependencies.Input('max-price' , 'value'),
     dash.dependencies.Input('Industry', 'value'),
     dash.dependencies.Input('vend-type', 'value'),
     dash.dependencies.Input('sales-rep' , 'value'),
    ]
)
def update_table_gen_sum1(selected1,selected_cat,selected_b_code,selected_min_price,selected_max_price,selected_indus,selected_v,selected_sr):
    
    dff = filter_data(df4,'ON_OFF',selected1)                                  #### Filter on the basis of product category
    dff = filter_data(dff,'Liquor.Category',selected_cat)                      #### Filter on the basis of product category
    dff = filter_data(dff,'Brand.Code',selected_b_code)                        #### Filter on the basis of brand code
    dff = filter_price(dff,selected_min_price,selected_max_price)              #### Filter on the basis of price range
    dff = filter_data(dff,'Industry',selected_indus)                           #### Filter on the basis of Industry
    dff = filter_radio(dff,'Deptt',selected_v)                                 #### Filter on the basis of vendor type
    dffx = filter_radio(df3,'Sales Rep name',selected_sr)                      #### Filter the selected Sales Reps only
    dff = dff[dff['Shop.Code'].isin(dffx['Shop Code'])]                        #### Filter on the shop codes present in selected vends
    #### Dropping extra redundant columns ####
    dff = dff.drop(labels = ['S.no.', 'Purchase.Order.No.', 'PO.Date', 'Supplier.Name', 'Supplier.Code',
                                'Transport.Permit.No.', 'TP.Status', 'TP.Date' ,  'Packing', 'WSP.per.Unit..INR.',
                                'WSP.per.Case..INR.', 'Excise.Revenue.per.Unit..INR.', 'MRP..INR.',
                                'WSP.per.Unit...No.of.Cases.Packing.Size','Excise.revenue.Per.Unit...No.Of.Cases..Packing.Size',
                                'Deptt','ON_OFF','Brand.Name','Brand.Code','Size..ml.'], axis=1)
    
    #### Applying group by on the basis of Month, Industry type,shop name ####
    dff = pd.pivot_table(dff, values='Quantity.Ordered', index=['Shop.Name','Industry'],columns=['Month'], aggfunc=np.sum)
    dff = dff.reset_index()
    x = df4['Month'].unique()
    x = np.append(x,'Shop.Name')
    x = x.tolist()
    dff = dff.sort_values(by = x ,ascending=False)
    return(generate_table(dff))

################   Body ends here   ######################

##  app callback to update the Sum2 to be tabulated  ##
################  Body starts here  ######################
@app.callback(
    dash.dependencies.Output('Table-Gen-Sum2', 'children'),
    [#dash.dependencies.Input('main-dataframe','children'),
     dash.dependencies.Input('Category', 'value'),
     dash.dependencies.Input('ON-OFF', 'value'),
     dash.dependencies.Input('Industry', 'value'),
     dash.dependencies.Input('min-price' , 'value'),
     dash.dependencies.Input('max-price' , 'value'),
    ]
)
def update_table_gen_sum2(selected_cat,selected_v_type,selected_indus,selected_min_price,selected_max_price):
    
    dff = filter_data(df4,'Liquor.Category',selected_cat)                  #### Filter on the basis of product category
    dff = filter_data(dff,'ON_OFF',selected_v_type)                        #### Filter on the basis of vendor type
    dff = filter_data(dff,'Industry',selected_indus)                       #### Filter on the basis of industry
    dff = filter_price(dff,selected_min_price,selected_max_price)          #### Filter on the basis of price range
    #### Dropping extra redundant columns ####
    dff = dff.drop(labels = ['S.no.', 'Purchase.Order.No.', 'PO.Date', 'Supplier.Name', 'Supplier.Code',
                                'Transport.Permit.No.', 'TP.Status', 'TP.Date' ,  'Packing', 'WSP.per.Unit..INR.',
                                'WSP.per.Case..INR.', 'Excise.Revenue.per.Unit..INR.', 'MRP..INR.',
                                'WSP.per.Unit...No.of.Cases.Packing.Size','Excise.revenue.Per.Unit...No.Of.Cases..Packing.Size',
                                'Deptt','ON_OFF','Brand.Code','Size..ml.','Shop.Name','Industry'], axis=1)
    
    #### Applying group by on the basis of Month, Industry type,shop name ####
    dff = pd.pivot_table(dff, values='Quantity.Ordered', index=['Brand.Name'],columns=['Month'], aggfunc=np.sum)
    x = df4['Month'].unique()
    x = x.tolist()
    dff = dff.reset_index()
    dff = dff.sort_values(by = x ,ascending=False)
    return(generate_table(dff))
################   Body ends here   ######################

#### app callback to update the file to be tabulated  ####
################  Body starts here  ######################

@app.callback(
    dash.dependencies.Output('Table Gen', 'children'),
    [#dash.dependencies.Input('main-dataframe','children'),
     dash.dependencies.Input('Industry', 'value'),
     dash.dependencies.Input('Category', 'value'),
     dash.dependencies.Input('Brand-Code', 'value'),
     dash.dependencies.Input('ON-OFF', 'value'),
     dash.dependencies.Input('vend-type', 'value'),
     dash.dependencies.Input('indicator-graphic', 'clickData'),
     dash.dependencies.Input('min-price' , 'value'),
     dash.dependencies.Input('max-price' , 'value'),
     dash.dependencies.Input('sales-rep' , 'value'),
    ]
)
def update_table_gen(selected_indus,selected_cat,selected_b_code,selected_v_type,selected_v,clicked_data,selected_min_price,selected_max_price,selected_sr):
    dff = filter_data(df4,'Industry',selected_indus)                  #### Filter on the basis of industry type
    dff = filter_data(dff,'Liquor.Category',selected_cat)             #### Filter on the basis of product category
    dff = filter_data(dff,'Brand.Code',selected_b_code)               #### Filter on the basis of brand code
    dff = filter_data(dff,'ON_OFF',selected_v_type)                   #### Filter on the basis of vendor type
    dff = filter_radio(dff,'Deptt',selected_v)                        #### Filter on the basis of vendor
    dff = filter_price(dff,selected_min_price,selected_max_price)     #### Filter on the basis of price range
    dffx = filter_radio(df3,'Sales Rep name',selected_sr)             #### Filter the selected Sales Reps only
    dff = dff[dff['Shop.Code'].isin(dffx['Shop Code'])]               #### Filter on the shop codes present in selected vends
    if clicked_data is not None:
        z = clicked_data['points'][0]['customdata']                   #### Getting clicked shop code
        dff = dff[dff['Shop.Code'] == z]                              #### Filter on the basis of clicked shop code
    else:
        dff = dff
        
     #### Dropping extra redundant columns ####
    dff = dff.drop(labels = ['S.no.', 'Purchase.Order.No.', 'PO.Date', 'Supplier.Name', 'Supplier.Code',
                                'Transport.Permit.No.', 'TP.Status', 'TP.Date' ,  'Packing', 'WSP.per.Unit..INR.',
                                'WSP.per.Case..INR.', 'Excise.Revenue.per.Unit..INR.', 'MRP..INR.','WSP.per.Unit...No.of.Cases.Packing.Size',
                                'Excise.revenue.Per.Unit...No.Of.Cases..Packing.Size', 'Deptt','ON_OFF'], axis=1)
    dff = dff.sort_values(by = ['Quantity.Ordered'], ascending=False)
    return(generate_table(dff))

################   Body ends here   ######################

##  app callback to update the groupby to be tabulated  ##
################  Body starts here  ######################
@app.callback(
    dash.dependencies.Output('Table-Gen-Groupby', 'children'),
    [#dash.dependencies.Input('main-dataframe','children'),
     dash.dependencies.Input('Industry', 'value'),
     dash.dependencies.Input('Category', 'value'),
     dash.dependencies.Input('Brand-Code', 'value'),
     dash.dependencies.Input('ON-OFF', 'value'),
     dash.dependencies.Input('vend-type', 'value'),
     dash.dependencies.Input('indicator-graphic', 'clickData'),
     dash.dependencies.Input('min-price' , 'value'),
     dash.dependencies.Input('max-price' , 'value'),
     dash.dependencies.Input('sales-rep' , 'value'),
    ]
)
def update_table_gen_groupby(selected_indus,selected_cat,selected_b_code,selected_v_type,selected_v,clicked_data,selected_min_price,selected_max_price,selected_sr):
    dff = filter_data(df4,'Industry',selected_indus)                  #### Filter on the basis of industry type
    dff = filter_data(dff,'Liquor.Category',selected_cat)             #### Filter on the basis of product category
    dff = filter_data(dff,'Brand.Code',selected_b_code)               #### Filter on the basis of brand code
    dff = filter_data(dff,'ON_OFF',selected_v_type)                   #### Filter on the basis of vendor type
    dff = filter_radio(dff,'Deptt',selected_v)                        #### Filter on the basis of vendor
    dff = filter_price(dff,selected_min_price,selected_max_price)     #### Filter on the basis of Price range
    dffx = filter_radio(df3,'Sales Rep name',selected_sr)             #### Filter the selected Sales Reps only
    dff = dff[dff['Shop.Code'].isin(dffx['Shop Code'])]               #### Filter on the shop codes present in selected vends
    if clicked_data is not None:
        z = clicked_data['points'][0]['customdata']                   #### Getting clicked shop code
        dff = dff[dff['Shop.Code'] == z]                              #### Filter on the basis of clicked shop code
    else:
        dff = dff
        
    #### Dropping extra redundant columns ####
    dff = dff.drop(labels = ['S.no.', 'Purchase.Order.No.', 'PO.Date', 'Supplier.Name', 'Supplier.Code',
                                'Transport.Permit.No.', 'TP.Status', 'TP.Date' ,  'Packing', 'WSP.per.Unit..INR.',
                                'WSP.per.Case..INR.', 'Excise.Revenue.per.Unit..INR.', 'MRP..INR.',
                                'WSP.per.Unit...No.of.Cases.Packing.Size','Excise.revenue.Per.Unit...No.Of.Cases..Packing.Size',
                                'Deptt','ON_OFF'], axis=1)
    
    #### Applying group by on the basis of Month, Industry type, Liquor category, Brand, size and shop name ####
    dff = dff.groupby(by=['Shop.Name','Industry','Liquor.Category','Brand.Name','Size..ml.','Month']).sum()
    dff = dff.reset_index()
    #dff = dff.sort_values(by = ['Quantity.Ordered'], ascending=False)
    return(generate_table(dff))

################   Body ends here   ######################




#### app callback to update the file to be downloaded ####
################  Body starts here  ######################

@app.callback(
    dash.dependencies.Output('download-link', 'href'),
    [#dash.dependencies.Input('main-dataframe','children'),
     dash.dependencies.Input('Industry', 'value'),
     dash.dependencies.Input('Category', 'value'),
     dash.dependencies.Input('Brand-Code', 'value'),
     dash.dependencies.Input('ON-OFF', 'value'),
     dash.dependencies.Input('vend-type', 'value'),
     dash.dependencies.Input('indicator-graphic', 'clickData'),
     dash.dependencies.Input('min-price' , 'value'),
     dash.dependencies.Input('max-price' , 'value'),
     dash.dependencies.Input('sales-rep' , 'value'),
    ]
)
def update_download_file(selected_indus,selected_cat,selected_b_code,selected_v_type,selected_v,clicked_data,selected_min_price,selected_max_price,selected_sr):
    dff = filter_data(df4,'Industry',selected_indus)                     #### Filter on the basis of industry type
    dff = filter_data(dff,'Liquor.Category',selected_cat)                #### Filter on the basis of product category
    dff = filter_data(dff,'Brand.Code',selected_b_code)                  #### Filter on the basis of brand code
    dff = filter_data(dff,'ON_OFF',selected_v_type)                      #### Filter on the basis of vendor type
    dff = filter_radio(dff,'Deptt',selected_v)                           #### Filter on the basis of vendor
    dff = filter_price(dff,selected_min_price,selected_max_price)        #### Filter on the basis of Price range
    dffx = filter_radio(df3,'Sales Rep name',selected_sr)                #### Filter the selected Sales Reps only
    dff = dff[dff['Shop.Code'].isin(dffx['Shop Code'])]                  #### Filter on the shop codes present in selected vends
    if clicked_data is not None:
        dff = dff[dff['Shop.Code'] == clicked_data['points'][0]['customdata']]
                                                                         #### Filter on the basis of clicked shop code
    else:
        dff = dff
    
    csv_string = dff.to_csv(index=False, encoding='utf-8')
    csv_string = "data:text/csv;charset=utf-8," + urllib.parse.quote(csv_string)
    return csv_string

################   Body ends here   ######################   

##  app callback to update the groupby to be downloaded ##
################  Body starts here  ######################

@app.callback(
    dash.dependencies.Output('download-link-groupby', 'href'),
    [#dash.dependencies.Input('main-dataframe','children'),
     dash.dependencies.Input('Industry', 'value'),
     dash.dependencies.Input('Category', 'value'),
     dash.dependencies.Input('Brand-Code', 'value'),
     dash.dependencies.Input('ON-OFF', 'value'),
     dash.dependencies.Input('vend-type', 'value'),
     dash.dependencies.Input('indicator-graphic', 'clickData'),
     dash.dependencies.Input('min-price' , 'value'),
     dash.dependencies.Input('max-price' , 'value'),
     dash.dependencies.Input('sales-rep' , 'value'),
    ]
)
def update_download_file_groupby(selected_indus,selected_cat,selected_b_code,selected_v_type,selected_v,clicked_data,selected_min_price,selected_max_price,selected_sr):
    dff = filter_data(df4,'Industry',selected_indus)                  #### Filter on the basis of industry type
    dff = filter_data(dff,'Liquor.Category',selected_cat)             #### Filter on the basis of product category
    dff = filter_data(dff,'Brand.Code',selected_b_code)               #### Filter on the basis of brand code
    dff = filter_data(dff,'ON_OFF',selected_v_type)                   #### Filter on the basis of vendor type
    dff = filter_radio(dff,'Deptt',selected_v)                        #### Filter on the basis of vendor
    dff = filter_price(dff,selected_min_price,selected_max_price)     #### Filter on the basis of Price range
    dffx = filter_radio(df3,'Sales Rep name',selected_sr)             #### Filter the selected Sales Reps only
    dff = dff[dff['Shop.Code'].isin(dffx['Shop Code'])]               #### Filter on the shop codes present in selected vends
    if clicked_data is not None:
        z = clicked_data['points'][0]['customdata']                   #### Getting clicked shop code
        dff = dff[dff['Shop.Code'] == z]                              #### Filter on the basis of clicked shop code
    else:
        dff = dff
        
    #### Dropping extra redundant columns ####
    dff = dff.drop(labels = ['S.no.', 'Purchase.Order.No.', 'PO.Date', 'Supplier.Name', 'Supplier.Code',
                                'Transport.Permit.No.', 'TP.Status', 'TP.Date' ,  'Packing', 'WSP.per.Unit..INR.',
                                'WSP.per.Case..INR.', 'Excise.Revenue.per.Unit..INR.', 'MRP..INR.',
                                'WSP.per.Unit...No.of.Cases.Packing.Size','Excise.revenue.Per.Unit...No.Of.Cases..Packing.Size',
                                'Deptt','ON_OFF'], axis=1)
    
    #### Applying group by on the basis of Month, Industry type, Liquor category, Brand, size and shop name ####
    dff = dff.groupby(by=['Month','Industry','Liquor.Category','Brand.Name','Size..ml.','Shop.Name']).sum()
    
    #### converting a groupby object into a dataframe by converting groupby addins as index ####
    dff = dff.reset_index()
    
    csv_string = dff.to_csv(index=False, encoding='utf-8')
    csv_string = "data:text/csv;charset=utf-8," + urllib.parse.quote(csv_string)
    return csv_string


################   Body ends here   ######################       


if __name__ == '__main__':
    app.run_server(port=8080, host='localhost')
```
